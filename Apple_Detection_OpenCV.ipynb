```python
#importing required libraries
import cv2 #computer vision package 
import xml.etree.ElementTree as et #xml file parsing package
from google.colab.patches import cv2_imshow #to show image
```

```python
image = cv2.imread("/content/data/apple_dataset/train/images/apple (267).jpg")
cv2_imshow(image)
```

```python
#path for image annotation
tree = et.parse("/content/data/apple_dataset/train/annotations/apple (267).xml")
root = tree.getroot()
bnd_list, name_list = [], []
for obj in root.findall('object'):
  for name in obj.findall('name'):
    name_list.append(name.text) #text label for image objects
  for bnd in obj.findall('bndbox'):
    #bounding box coordinates for image objects
      xmax = int(bnd.find('xmax').text)
      ymax = int(bnd.find('ymax').text)
      xmin = int(bnd.find('xmin').text)
      ymin = int(bnd.find('ymin').text)
      frame = [xmin,ymin,xmax,ymax]
      bnd_list.append(frame)

#reading image using cv2.imread
image = cv2.imread("/content/data/apple_dataset/train/images/apple (267).jpg")
for i, j in zip(range(len(bnd_list)), name_list):
  (x, y) = (bnd_list[i][0], bnd_list[i][1])
  (w, h) = (bnd_list[i][2], bnd_list[i][3])
  cv2.rectangle(image, (x, y), (x + w, y + h), (255, 0, 0), 1)
  cv2.putText(image, j, (x, y - 5), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255,0,0), 1)

cv2_imshow(image)
```
